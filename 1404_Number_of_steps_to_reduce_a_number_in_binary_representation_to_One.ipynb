#### Solution 1

```python
class Solution:
    def numSteps(self, s: str) -> int:
        c = 0
        s = int(s, 2) # converting binary to integer
        while(s!=1):
            # s = int(s,2)
            if s%2==0:
                c = c+1
                s = s//2
                print(s)
            else:
                s = s+1
                c = c+1
                print(s)
        return c
            
            
```
