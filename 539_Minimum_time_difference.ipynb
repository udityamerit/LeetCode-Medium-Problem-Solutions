#### Solution 1

```python
class Solution:
    def findMinDifference(self, timePoints: List[str]) -> int:
        MIn = []
        for time in timePoints:
            h, m = map(int, time.split(':'))
            MIn.append(h*60+m)
        
        MIn.sort()

        min_diff = float('inf')
        for i in range(len(MIn)-1):
            min_diff = min(min_diff, MIn[i+1] - MIn[i])

        min_diff = min(min_diff, 1440-MIn[-1]+MIn[0])

        return min_diff
```
