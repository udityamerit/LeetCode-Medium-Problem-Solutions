```python
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        # l = []
        # for i in range(len(nums)):
        #     l = sum(nums[i:])
        #     print(l)
        #     if l%k==0:
        #         return True
        
        # return False
        map = {0: -1}  
        sum = 0

        for i in range(len(nums)):
            sum += nums[i]
            rem = sum % k

            if rem in map:
                if i - map[rem] > 1:
                    return True
            else:
                map[rem] = i

        return False
        
```
