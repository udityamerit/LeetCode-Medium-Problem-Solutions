#### Solution 1

```python

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:

        # median for even numbers: [(n/2)term+{(n/2)+1}term]/2
        # median for odd numbers: [(n+1)/2]term

        num3=nums1+nums2
        num3.sort()
        m=len(num3)
        if m%2==1:
            return num3[m//2]
        else:
            mid=m//2
            return (num3[mid-1]+num3[mid])/2
```
