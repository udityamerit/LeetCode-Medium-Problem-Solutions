```python
class Solution:
    def dividePlayers(self, skill: List[int]) -> int:
        l =[]
        skill.sort()
        m = skill[0]
        n = skill[-1]
        k=[skill[i] for i in range(len(skill)//2)]
        u = [skill[-i] for i in range(1,(len(skill)//2)+1 )]
        print(k)
        print(u)
        c = []
        for i in range(len(k)):
            if k[i]+u[i]==m+n:
                c.append([k[i],u[i]])

        chemistry = 0
        if len(skill)/2 ==len(c):
            for i in range(len(c)):
                chemistry = chemistry+(c[i][0]*c[i][1])
            return chemistry  
        else:
            return -1     
```
