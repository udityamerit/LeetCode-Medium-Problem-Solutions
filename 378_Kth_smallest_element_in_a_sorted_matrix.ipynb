```python
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        l = []
        for i in range(len(matrix)):
            for j in range(len(matrix)):
                l.append(matrix[i][j])
        l.sort()
        print(l)
        return l[k-1]
```
