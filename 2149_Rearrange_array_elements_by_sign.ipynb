#### Solution 1

```python
class Solution:
    def rearrangeArray(self, nums: List[int]) -> List[int]:
        l = [i for i in nums if i>0]
        k = [i for i in nums if i<0]
        print(l)
        print(k)
        m = []
        j = 0
        c = 0
        for i in range(len(nums)):
            if i%2==0:
                
                m.append(l[j])
                j = j+1
            else:
                m.append(k[c])
                c = c+1
        return m
```

```python
class Solution:
    def rearrangeArray(self, nums: List[int]) -> List[int]:
        l = [i for i in nums if i>0]
        k = [i for i in nums if i<0]
        print(l)
        print(k)
        m = []
        
        for i in range(len(l)):
            m.append(l[i])
            m.append(k[i])
        return m
```
