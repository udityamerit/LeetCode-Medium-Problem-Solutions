#### Solution 1

```python
import math
import sys
class Solution:
    def trailingZeroes(self, n: int) -> int:
        sys.set_int_max_str_digits(0)
        c = math.factorial(n)
        # print(c)       
        l = [i for i in str(c)]
        m = 0
        for i in range(1,len(l)+1):
            if l[-i]=='0':
                m +=1
            if l[-i]!= '0':
                break
        return m
```
