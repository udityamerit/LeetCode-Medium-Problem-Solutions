#### Solution 1

```python
class Solution:
    def maxBottlesDrunk(self, numBottles: int, numExchange: int) -> int:
        o = numBottles
        a = o
        while(o>=numExchange):
            x=1
            a=a+x
            o=o-numExchange+1
            numExchange=numExchange+1
        return a   
        
        
```
