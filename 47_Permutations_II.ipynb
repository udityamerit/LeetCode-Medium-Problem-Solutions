#### Solution 1 in python

```python
class Solution:
    def permuteUnique(self, nums: List[int]) -> List[List[int]]:
        l = []
        p = permutations(nums)
        for j in list(p):
           
            l.append(list(j))
      
        c = []
        for i in l:
            if i not in c:
                c.append(i)
        # print(c)
        return c
```
