```python
class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        # c = 0
        arr = []
        for i in range(len(nums)):
            for j in range(i,len(nums)):
                if i!=j:
                    if(abs(nums[i]-nums[j])==k):
                        if ((nums[i], nums[j])) not in arr:

                            arr.append([nums[i], nums[j]])
                
        print(arr)
        a = []
        for i in arr:
            i.sort()
            if i not in a:

                a.append(i)
        print(a)
       
        return len(a)
        
            
```
