#### Solution 1

```python
class Solution:
    def kthFactor(self, n: int, k: int) -> int:
        l = [ ]
        for i in range(1,n+1):
            if n%i==0:
                l.append(i)
        print(l)
        
        if len(l)>=k:
            return l[k-1]
        else:
            return -1
```
