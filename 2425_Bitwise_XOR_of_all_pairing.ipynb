```python
class Solution:
    def xorAllNums(self, nums1: List[int], nums2: List[int]) -> int:
        # num3 = []
        
        # for i in nums1:
        #     for j in nums2:
        #         num3.append(i^j)
        # print(num3)
        # val = 0
        # for i in num3
        #     val = val^num3[i]

        # return val

        c1 = len(nums1)
        c2 = len(nums2)
        x1 = 0
        x2 = 0
        if c1 % 2 != 0:
            for i in nums2:
                x2 ^= i
        if c2 % 2 != 0:
            for i in nums1:
                x1 ^= i
        return x1 ^ x2
```
