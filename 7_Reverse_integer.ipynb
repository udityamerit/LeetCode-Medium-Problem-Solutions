#### Solution 1

```python
class Solution:
    def reverse(self, x: int) -> int:
        res = 0
        
        
            
        if x>=0:

            while x>0:
                r = x%10
                res = res*10 +r
                x = x//10

            if res > 2 ** 31 - 1 or res < -2 ** 31:
                return 0
            else:
                return res
        
        if x<0:

            x = str(x)
            x = [s for s in x]
            x.remove(x[0])
            x = int("".join(x))
            
            while x>0:
                r = x%10
                res = res*10 +r
                x = x//10

            if res > 2 ** 31 - 1 or res < -2 ** 31:
                return 0
            else:
                return -(res)
        
```
