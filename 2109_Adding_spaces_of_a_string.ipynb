#### Solution 1 in Python

```python
class Solution:
    def addSpaces(self, s: str, spaces: List[int]) -> str:
        result = []
        prev_index = 0
        
        for space in spaces:
            result.append(s[prev_index:space])  
            result.append(' ')                 
            prev_index = space                
            # print(result)
        result.append(s[prev_index:])         
        return ''.join(result)
```
