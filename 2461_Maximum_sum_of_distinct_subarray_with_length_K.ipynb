#### Solution 1 in Python

```python
from collections import Counter
from typing import List

class Solution:
    def maximumSubarraySum(self, nums: List[int], k: int) -> int:
        n = len(nums)
        if n < k:
            return 0
        
        
        current_sum = 0
        count = Counter()
        max_sum = 0
        
        for i in range(n):
          
            count[nums[i]] += 1
            current_sum += nums[i]
            
            if i >= k:
                count[nums[i - k]] -= 1
                if count[nums[i - k]] == 0:
                    del count[nums[i - k]]
                current_sum -= nums[i - k]
            
           
            if i >= k - 1 and len(count) == k:
                max_sum = max(max_sum, current_sum)
        
        return max_sum
```
