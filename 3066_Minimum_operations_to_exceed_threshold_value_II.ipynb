```python
class Solution:
    def minOperations(self, nums: List[int], k: int) -> int:
        # this solution is supported to the 600 test case
        c = 0
        # print(nums)
        for i in range(len(nums)):
            if min(nums)<k:
                nums.sort()
                a = nums[0]
                b = nums[1]
                nums.remove(a)
                nums.remove(b)
                nums.append(min(a,b)*2 + max(a,b))
                c = c+1
            if min(nums)>=k:
                break
        # print(nums)
        # print(c)
        return c
```
