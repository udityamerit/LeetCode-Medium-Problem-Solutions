#### Solution 1

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        ans = []
        if len(nums)>=2:
            for i in range(len(nums)):
                if nums.count(nums[i])>len(nums)/3:
                    ans.append(nums[i])
            return list(set(ans))

        if len(nums)==1:
            return nums
```
