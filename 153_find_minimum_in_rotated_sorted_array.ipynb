#### Solution 1

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left = 0
        right = len(nums) - 1

        while left < right:
            middle = (left + right) // 2
            if nums[middle] >= nums[0] and nums[right] < nums[0]:
                left = middle + 1
            else:
                right = middle
        return nums[left]
```

#### Solution 2

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        return min(nums)
    
    
```
