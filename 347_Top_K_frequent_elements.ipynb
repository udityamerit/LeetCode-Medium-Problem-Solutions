```python
from collections import Counter
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        nums = Counter(nums)
        l = []
        k1 = []
        for key, value in nums.items():
            l.append(value)
        l.sort(reverse=True)
        print(l)
        # i = 0
        for i in range(k):
            for key,value in nums.items():
                if nums.get(key)==l[i]:
                    k1.append(key)
                    # i = i+1
        return set(k1)
```
