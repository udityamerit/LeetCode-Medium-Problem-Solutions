```python
class Solution:
    def smallestNumber(self, num: int) -> int:
        if num<0:
            l = [i for i in str(num)]
            del l[0]
            # print(l)
            return -int(''.join(sorted(l)[::-1]))
        else:
            # # l = [i for i in str(num)]
            # c = list(permutations((str(num))))
            # # print(c)
            # c = [list(i) for i in c]
            # d = [int(''.join(perm)) for perm in c if perm[0] != '0']
            # return (min(d))
            l = sorted(str(num))
            if l[0] == '0':
                for i in range(1, len(l)):
                    if l[i] != '0':
                        l[0], l[i] = l[i], '0'
                        break
            return int(''.join(l))
```
