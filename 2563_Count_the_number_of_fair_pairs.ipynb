#### Solution 1

```python
class Solution:
    def countFairPairs(self, nums: List[int], lower: int, upper: int) -> int:
        nums.sort()  
        count = 0
                
        def countPairs(target):
            left, right = 0, len(nums) - 1
            total = 0
            while left < right:
                if nums[left] + nums[right] <= target:
                    total += right - left  
                    left += 1
                else:
                    right -= 1
            return total

        count = countPairs(upper) - countPairs(lower - 1)
        return count
```
