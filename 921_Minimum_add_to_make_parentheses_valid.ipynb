#### Solution 1

```python
class Solution:
    def minAddToMakeValid(self, s: str) -> int:
      c = 0
      d = 0
      for i in s:

            if i =='(':
                c = c+1
            elif i==')' and c>0:
                c  = c-1
            else:
                d = d+1
        
      return c+d
```
