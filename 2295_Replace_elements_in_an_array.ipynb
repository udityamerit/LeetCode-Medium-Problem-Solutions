#### Solution 1 in Python

```python
from typing import List

class Solution:
    def arrayChange(self, nums: List[int], operations: List[List[int]]) -> List[int]:
        value_map = {}  
        for a, b in operations:
            if a in value_map:
                value_map[b] = value_map.pop(a)
            else:
                value_map[b] = a
        
        
        final_map = {v: k for k, v in value_map.items()}
        
        return [final_map.get(num, num) for num in nums]
```
