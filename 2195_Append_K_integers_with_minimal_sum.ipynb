#### Solution 1 in Python

```python
class Solution:
    def minimalKSum(self, nums: List[int], k: int) -> int:
        nums=list(set(nums))
        nums.sort()
        # print(nums)
        n=len(nums)
        sums=0
        for i in range(n):
            if nums[i]<=k:
                k+=1
                sums+=nums[i]
                # print(k,nums[i],sums)
        res=k*(k+1)//2-sums
        return res
```
