#### Solution 1

```python
class Solution: 
    def nextGreaterElement(self, n: int) -> int:
        x = [i for i in str(n)]
        list1 = list(set([int("". join(i)) for i in permutations(x, len(x)) if int("". join(i))> n])) 
        list1.sort()
        
        if len(list1)!=0 and list1[0]<2**31:
            return list1[0]
        else:
            return -1
        
```
