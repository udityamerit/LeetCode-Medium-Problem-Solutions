#### Solution 1 in Python

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        for i in range(1, k+1):
            c = nums[-1]
            # print(c)
            nums.insert(0, c)
            # print(nums)
            del nums[-1]
        # print(nums)
        return nums
```
